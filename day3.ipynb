```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("water_usage_prediction.csv")
```

```python
df.head()
```

```python
df.info()
```

**talking about linear regression**


```python
df.tail()
```

```python
X = df[['Year']]
y = df['Annual Rainfall (mm)']

# Train linear regression model
model = LinearRegression()
model.fit(X, y)

```

```python
year_to_predict = np.array([[2026]])
predicted_rainfall = model.predict(year_to_predict)
```

```python
print(f"Predicted Annual Rainfall for 2026: {predicted_rainfall[0]:.2f} mm")
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(year_to_predict, predicted_rainfall, color='green', marker='x', s=100, label='Prediction (2026)')
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.title('Annual Rainfall Prediction using Linear Regression')
plt.legend()
plt.show()
```
