```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("water_usage_prediction.csv")
```

```python
df.head()
```

```python
df.info()
```

**talking about linear regression**


```python
df.tail()
```

```python
X = df[['Year']]
y = df['Annual Rainfall (mm)']

# Train linear regression model
model = LinearRegression()
model.fit(X, y)

```

```python
year_to_predict = np.array([[2026]])
predicted_rainfall = model.predict(year_to_predict)
```

```python
print(f"Predicted Annual Rainfall for 2026: {predicted_rainfall[0]:.2f} mm")
```

```python
plt.bar(X['Year'], y, color='blue', label='Actual Data')

# Overlay the regression line
plt.plot(X['Year'], model.predict(X), color='red', label='Regression Line')

# Highlight the predicted rainfall for 2026
plt.bar(2026, predicted_rainfall[0], color='green', label='Prediction (2026)')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.title('Annual Rainfall Prediction using Linear Regression')
plt.legend()
plt.show()
```

```python
average_rainfall = df['Annual Rainfall (mm)'].mean()

print(f"Average Annual Rainfall: {average_rainfall:.2f} mm")
```
