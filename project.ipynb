```python
df.drop(columns=["Link"],inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['District_Name'] = label_encoder.fit_transform(df['District_Name'])
df['Soil_color'] = label_encoder.fit_transform(df['Soil_color'])
df['Fertilizer'] = label_encoder.fit_transform(df['Fertilizer'])
df['Crop'] = label_encoder.fit_transform(df['Crop'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns=["Fertilizer_Usage"])  # Features
y = df["Fertilizer_Usage"]  # Target variable
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Step 5: Normalize the data to improve model performance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)


```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(ytest, ypred)
mse = mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)

# Print results
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating index for x-axis
indices = np.arange(len(ytest))

# Bar width
width = 0.4  

plt.figure(figsize=(12, 6))

# Plot actual values
plt.bar(indices - width/2, ytest, width=width, color='blue', label="Actual Fertilizer Usage")

# Plot predicted values
plt.bar(indices + width/2, ypred, width=width, color='red', label="Predicted Fertilizer Usage")

# Labels and title
plt.xlabel("Test Sample Index")
plt.ylabel("Fertilizer Usage")
plt.title("Actual vs Predicted Fertilizer Usage")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Scatter plot for actual vs predicted
plt.scatter(ytest, ypred, color='blue', label="Predicted vs Actual")

# Diagonal line for perfect predictions
plt.plot([min(ytest), max(ytest)], [min(ytest), max(ytest)], color='red', linestyle="dashed", label="Perfect Fit Line")

# Labels and title
plt.xlabel("Actual Fertilizer Usage")
plt.ylabel("Predicted Fertilizer Usage")
plt.title("Actual vs Predicted Fertilizer Usage (Linear Regression)")
plt.legend()
plt.show()
```
